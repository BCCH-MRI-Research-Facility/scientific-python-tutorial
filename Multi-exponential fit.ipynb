```python
def t2decay(t,A,B,T2myelin,T2ieh2O):
    return A*np.exp(-t/T2myelin) + B*np.exp(-t/T2ieh2O)
```

```python
xdata = np.linspace(0,32,32)
```

```python
ydataraw = t2decay(xdata,10,90,15,45)
```

```python
ydata = ydataraw + np.random.random(32)
```

```python
f,ax = plt.subplots()
```

```python
ax.scatter(xdata,ydata)
```

```python
from scipy.optimize import curve_fit
```

```python
popt,pcov = curve_fit(t2decay,xdata,ydata,bounds=([0,0,10,40],[np.inf,np.inf,20,50]),p0=[10,90,15,45])
```

```python
popt
```

```python
pcov
```

```python
fitdata = t2decay(xdata,*popt)
```

```python
plt.scatter(xdata,fitdata)
```
